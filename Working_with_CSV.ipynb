```python
import pandas as pd

# Read the data into a DataFrame, selecting only the 'enrollee_id' column
df_enrollee_id = pd.read_csv('aug_train.csv', usecols=['gender'])

# Convert the single-column DataFrame to a Series using .squeeze()
# .squeeze() is still available as a DataFrame method.
enrollee_id_series = df_enrollee_id['gender'].squeeze()

# Display the series (or print its head/type to confirm)
print(enrollee_id_series.head())
print(type(enrollee_id_series))
```
