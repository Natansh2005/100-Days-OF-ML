```python
for i in range(1,501):
  response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=0c98e0847a6f88ba31c49c8ff95a9c81&language=en-US&page={}'.format(i))
  json_response = response.json()
  if 'results' not in json_response:
    print(f"Error: 'results' key not found in response for page {i}. Full response:")
    print(json_response)
    break # Stop the loop if 'results' is missing
  temp_dr = pd.DataFrame(json_response['results']).head(2)[['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']]
  df = pd.concat([df, temp_dr], ignore_index=True)
```
