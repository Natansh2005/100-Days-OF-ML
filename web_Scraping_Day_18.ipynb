```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

final=pd.DataFrame()

for j in range(1,11):
  url='https://internshala.com/internships/engineering-internship/page-{}'.format(j)
  headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
  webpage = requests.get(url, headers=headers).text

  soup=BeautifulSoup(webpage, 'lxml')
  company=soup.find_all('div', class_='individual_internship')

  name=[]
  hiring=[]
  stipend=[]
  loc=[]
  old=[]

  for i in company:
    job_title_tag = i.find('h2', class_='job-internship-name')
    name.append(job_title_tag.text.strip() if job_title_tag else "N/A")

    job_title_tag1 = i.find('div', class_='actively-hiring-badge')
    hiring.append(job_title_tag1.text.strip() if job_title_tag1 else "N/A")

    job_title_tag1 = i.find('span', class_='stipend')
    stipend.append(job_title_tag1.text.strip() if job_title_tag1 else "N/A")

    # Extract location
    location_tag = i.find('div', class_='locations')
    loc.append(location_tag.find('a').text.strip() if location_tag and location_tag.find('a') else "N/A")

    # Extract posting age
    old_tag = i.find('div', class_='status-info')
    old.append(old_tag.find('span').text.strip() if old_tag and old_tag.find('span') else "N/A")

  d={'name':name, 'hiring':hiring, 'stipend':stipend, 'location':loc, 'posted_ago':old}
  df=pd.DataFrame(d)

  final = pd.concat([final, df], ignore_index=True)
```
